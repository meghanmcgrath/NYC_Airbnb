Question 6 - Categorical plot

```python
sns.histplot(df_nyc_3['price_per_night'], bins = 10, kde = True)
```

```python
df_nyc_3.loc[df_nyc_3['price_per_night'] < 100, 'Price category'] = 'Low price' 
df_nyc_3.loc[(df_nyc_3['price_per_night'] >= 100) & (df_nyc_3['price_per_night'] < 200),'Price category'] = 'Middle price'
df_nyc_3.loc[df_nyc_3['price_per_night'] >= 200, 'Price category'] = 'High price' 
df_nyc_3['Price category'].value_counts(dropna = False)

# Create a categorical plot in seaborn using the price categories created above
sns.set(style="ticks")
g = sns.catplot(x="number_of_reviews", y="price_per_night", hue="Price category", data=df_nyc_3)
```
